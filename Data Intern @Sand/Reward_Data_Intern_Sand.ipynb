```python
#Import Necessary Library
import pandas as pd
```

```python
#Load Dataset
df=pd.read_excel('/content/_Data Sets Reward_Program_Assignment_Input_v6 - TA.xlsx', sheet_name ='Mentorship_Sessions')
# Display initial data
df.head()
```

```python
# Check data types
df.dtypes
```

```python
#Check missing values
df.isnull().sum()
```

```python
# Drop unwanted columns but keep others
df = df.drop(['Points_Awarded'], axis=1)
```

```python
# Drop the first column since it's not needed
df= df.drop(df.columns[0], axis=1)
```

```python
# Remove rows with NaN values
df = df.dropna()
```

```python
#Confirm there is no nulls
df.isnull().sum()
```

```python
#Check the number of duplicates
df.duplicated().sum()
```

```python
#Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
# Filter out specific rows based on conditions
df = df[df['Job_Info_Completed'] != 'Maybe']
```

```python
#Have a look at the dataset once more
df
```

```python
#Change the data types of these 3 columns
df['Mentor_ID'] = df['Mentor_ID'].astype(int)
df['Session_Number'] = df['Session_Number'].astype(int)
df['Session_Duration_Min'] = df['Session_Duration_Min'].astype(int)
```

```python
#Save Cleaned Data to an excel file
df.to_excel('clean_data.xlsx', index=False)
```
